```python
class Person:
    def __init__(self, name, age):
        self.name = name
        self.age = age

    def introduce(self):
        print(f"Hello, my name is {self.name} and I am {self.age} years old.")

class Employee:
    def __init__(self, employee_id, department):
        self.employee_id = employee_id
        self.department = department

    def work(self):
        print(f"Employee {self.employee_id} is working in {self.department} department.")

class Manager(Person, Employee):
    def __init__(self, name, age, employee_id, department):
        Person.__init__(self, name, age)
        Employee.__init__(self, employee_id, department)

    def manage(self):
        print(f"Manager {self.name} is managing the team in the {self.department} department.")

# Create Manager instance and test all methods
manager = Manager("John", 35, "M123", "HR")
manager.introduce()
manager.work()
manager.manage()
```

```python
class Greeting:
    def greet(self, *args):
        if len(args) == 0:
            print("Hello!")
        elif len(args) == 1:
            print(f"Hello, {args[0]}!")
        elif len(args) == 2:
            print(f"{args[1]}, {args[0]}!")
        else:
            print("Invalid number of arguments.")

g = Greeting()
g.greet()
g.greet("John")
g.greet("John", "Good morning")
```

```python
class SecuritySystem:
    def __init__(self, pin):
        self.__pin = pin  # Private attribute

    def validate_pin(self, pin):
        if pin == self.__pin:
            print("PIN validated successfully.")
            return True
        else:
            print("Invalid PIN.")
            return False

    def change_pin(self, old_pin, new_pin):
        if old_pin != self.__pin:
            print("Incorrect old PIN. PIN not changed.")
            return
        if not (new_pin.isdigit() and len(new_pin) == 4):
            print("New PIN must be a 4-digit number.")
            return
        self.__pin = new_pin
        print("PIN changed successfully.")

# Usage
security = SecuritySystem("1234")
security.validate_pin("1234")     # Valid
security.validate_pin("0000")     # Invalid
security.change_pin("0000", "5678")  # Wrong old PIN
security.change_pin("1234", "56")     # Invalid new PIN
security.change_pin("1234", "5678")  # Correct change
```

```python
from abc import ABC, abstractmethod
import math

class Shape(ABC):
    @abstractmethod
    def area(self):
        pass

class Circle(Shape):
    def __init__(self, radius):
        self.radius = radius

    def area(self):
        return round(math.pi * self.radius ** 2, 1)

class Rectangle(Shape):
    def __init__(self, width, height):
        self.width = width
        self.height = height

    def area(self):
        return self.width * self.height

# Create instances
circle = Circle(5)
rectangle = Rectangle(6, 4)

print("Area of Circle:", circle.area())
print("Area of Rectangle:", rectangle.area())
```

```python
class Notification:
    def send(self):
        print("Sending notification")

class EmailNotification(Notification):
    def send(self):
        print("Sending email")

class SMSNotification(Notification):
    def send(self):
        print("Sending SMS")

# Test
email = EmailNotification()
sms = SMSNotification()

email.send()
sms.send()
```
